## Using the python data science library pandas

Today we will introduce a new python library, pandas.  Pandas is a powerful library used in data science and statistics.  The core data structure of pandas is a `dataframe` which is your data set organized into rows and columns of data, much like a "sheet" in a spreadsheet program.  Often, pandas can process much larger data sets than traditional spreadsheet programs, and since pandas is a python library, it can integrate with the rest of the python ecosystem.  

In numpy arrays, the rows and columns are labeled with indexes (numbers) and to access a particular element, row, or column, you use the index numbers.  In pandas dataframes, the rows and columns are labled with names, which could be numbers or words.  To access a particular column in a pandas dataframe, the syntax is `dataframe['Column Name']`.  You can also create a new column in your dataframe in the same way.  You can also create a new column in the data frame by doing a calcuation with an existing column.  All of these examples are shown in the code block below.

To access a particular elements of a data frame, there are two ways to do it.  The `dataframe_name.iloc[row, column]` command works just like numpy, where the *row* and *column* values are the index numbers.  The more clever way to access data is `dataframe.loc['row name']['column name']`.  This way you don't have to count out the row or column index you need, you can just look it up by the name.  An example of this is shown in the code block below.

It is not always a clear cut choice whether a pandas dataframe or a numpy array is the better choice for a dataset.  Many times either will work with different pros and cons.   A good rule of thumb is that if a data set is all numerical data, a numpy array is the better choice.  If you will have different types of data, such as molecule names, attributes, and numerical data, a pandas dataframe is the better option.

## Background

In this experiment, we will connect the visible absorption spectrum of several polymethine dyes to elementary quantum mechanics, namely the one-dimensional particle in a box. This absorption corresponds to an electronic transition in the molecule between the highest occupied electronic energy level (HOMO) and the lowest unoccupied electronic energy level (LUMO). The difference between these two electronic energy levels will be estimated from the photon wavelength at the maximum absorbance at the longest wavelength. This energy separation will be compared to the energy separation predicted by using a simple "free electron" model calculation.

The absorption spectra of polymethine dyes arise from the electronic transitions involving the π electrons along the polymethine chain. The structure of one of the dyes is shown in Figure 1. The π electrons are delocalized between the two nitrogen atoms of the chain as shown below.

![Figure1 1](images/Lab4_fig1.png)

Each carbon atom in the chain and each nitrogen at the terminus are bonded to three other atoms through $\sigma$ bonding, which leaves one valence electron per carbon that may be involved in $\pi$ bonding. In addition, the three remaining valence electrons on the two nitrogen atoms contribute to the $\pi$ system. The *free electron model* assumes that the $\pi$ electrons are subject to a constant potential as they move along the polymethine chain as indicated in Figure 1. Furthermore, it is assumed that the potential increases very sharply in the region of the nitrogen atom and that the electrons behave as if they were confined to a "one–dimensional box." Finally, we assume that electron–electron repulsion is negligible.

In this lab, we will model the $\pi$ electron system as a particle in a box.  The energy of one of the electrons in the $\pi$ system is 
$$ E_n = \frac{n^2 h^2}{8mL^2} $$

We remember that experimentally, we can not measure the energy of any particular quantum state.  In spectroscopy experiments, we measure the *difference* in energy between two quantum states.  If N is the number of $\pi$ electrons in the conjugated chain, then for the ground state, there are $N/2$ occupied energy levels.  So, for example, in a system with 12 $\pi$ electrons, there would be 6 occupied energy levels.  The n=6 energy level would be the highest occupied molecular orbital (HOMO) and the n=7 level would be the lowest unoccupied molecular orbital (LUMO).  The HOMO-LUMO gap therefore represents the smallest change in energy which could occur between quantum states; this would correspond to the longest wavelength absorbance by the molecule in a spectroscopy experiment.  This $\Delta$E of the HOMO-LUMO gap is equal to the energy of the photon.

$$ E_{LUMO} - E_{HOMO} = E_{photon} $$

Since we can model the energy of the HOMO and the LUMO with the particle in a box equation, that gives us:

$$ (n_L^2 - n_H^2) \frac{h^2}{8mL^2} = \frac{hc}{\lambda} $$

We can estimate this energy with the free electron model if we can estimate the length of the box.  We will denote theh number of $\pi$ electrons in the box as p; this is the the number of carbon atoms in a polymethine chain plus the 3 electrons from the nitrogens.  We assume that L is the length of the chain between nitrogen atoms plus one bond distance on each side; therefore L can be found by 

$$ L = (p)*\ell $$ 

where $\ell$ is the bond length between the carbon atoms on the chain.  A reasonable value for $\ell$ is 139 pm.

## Part 2: Improving this results with an empirical correction

Our calculated $\lambda$ values do not agree particularly well with our experimental values, particularly for the 1,1-2,2 dye series.  When a model does not match observations, you should consider the assumptions of the model and identify where it can be improved.  One assumption we are making in using particle in a box for the model of the $\pi$ electrons is that the large aromatic rings create an infinite potential barrier.  This of course isn't actually true; the potential barrier is large but not infinite, so there can be some penetration beyond the barrier.  Thus, the length of our box is larger than we have estimated above.  

One popular, but not always good, way to improve a model is to add an empirical correction and then somehow use data that you have to fit a value to that correction factor.  In our case, we will introduction a correction factor, $\alpha$, to the length of the box, such that

$$ L = (p + \alpha) \ell. $$

To find the value of $\alpha$ we will use the measured value of wavelength for one of the dyes and work backwards to find the $\Delta$E, then the L, and then the $\alpha$ value.
