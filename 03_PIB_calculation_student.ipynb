## Background: the Particle in a Box Model

The particle in a box is a model problem in quantum mechanics that imagines there is a particle confined by infinitely high potential walls in a box of length $L$.
The wavefunction $\Psi(x)$ for this system is given by the equation:

$$\Psi(x) = \sqrt{\frac{2}{L}} \sin \bigg(\frac{n \pi x}{L} \bigg)$$

where $n = 1, 2, 3, ...$ is a quantum number.

**General Programming Tips**

a. In this lab you will need to plot someting in 3 dimensions.
Below is some example code of how to plot $z(x,y) = x^2 - y^2$.

b. You may also want to use the sympy solve fucntion with a restricted domain, this is best done with:

`sp.solveset(expression, variable, sp.Interval(start,end))`

c. Remember that when you want to change a symbolic expression into a numerical function, you use `sp.lambdify`.  If you want to substitute in a few values but have it remain a symbolic function, you use `sp.subs`.

```python
# 3D plotting example code

# Define our function
x, y = sp.symbols('x y')
z = x**2 - y**2

z_num = sp.lambdify([x, y], z, 'numpy')

x_vals = np.linspace(-1, 1, 50)
y_vals = np.linspace(-1, 1, 50)

# meshgrid is just generates intermediate inputs we need to find the 'matrix'
# of z values Z
X, Y = np.meshgrid(x_vals, y_vals)
Z = z_num(X, Y)

# The actual call for 3D plotting
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')

# the fourth argument here (100) is called the number of levels
# and is used to determine the numbers and positions of the
# contour lines / regions.
ax.contour3D(X, Y, Z, 100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
