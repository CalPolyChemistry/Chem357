## Background: Rovibrational Spectroscopy

A high-resolution spectrum of CO contains information about both vibrational and rotational energy levels of the molecule. The vibrational energy levels are described by the vibrational quantum number ν (ν = 0, 1, 2, ...). In this experiment, we will investigate both the ν = 0 to ν = 1 transition and the ν = 0 to ν = 2 transition. In addition to changing vibrational state, the molecule also changes rotational state, described by the rotational quantum number J (J = 0, 1, 2, ...). The rotation-vibration energy (in cm$^{-1}$) of a diatomic molecule is given by:

$$
E(J, \nu) = \bigg( \nu + \frac{1}{2} \bigg)\bar{\nu}
- \bigg( \nu + \frac{1}{2} \bigg)^2 \chi_e \bar{\nu}
+ J \big(J + 1 \big) B_e
- J \big(J + 1 \big) \bigg( \nu + \frac{1}{2} \bigg)\alpha_e
$$

The $( ν + \frac{1}{2} )\bar{\nu}$ term is the contribution due to perfect harmonic oscillation. However, a real molecule is not a perfect oscillator, so the correction term $(\nu + 1/2)^2 \chi_e \bar{\nu}$ is subtracted to account for anharmonic effects. The $J(J+1) B_e$ term is the rotational contribution to the energy assuming the molecule is a rigid rotor with fixed internuclear separation. However, the molecule is vibrating and the internuclear distance is not quite the same in the two different vibrational states. To account for this, another correction term, $J(J+1)(\nu +1/2) \alpha_e$, the rotation-vibration interaction term, is included. 

Note that although many of these values are in principle constants, because you will be using fitted data for this lab, you may get different values for the Coupling constant, the rotational contant, and the equilibrium bond length based on what vibrational transition data you are using. This is reflected in the final data table at the end of the lab.

Selection rules for rotation-vibration IR transitions only permit certain transitions between energy levels. The basic selection rules are Δν = ±1 and ΔJ = ±1. We are dealing with Δν = +1 in our recorded spectrum (why not Δν = -1?). The absorption lines with ΔJ = +1 produce the R branch of the spectrum and those with ΔJ = -1 produce the P branch. Figure 1 shows the origin of the R and P branch lines.

![image](images/Lab5_figure1.png)

The vertical lines represent allowed infrared rotation-vibration transitions for the v = 0
to v = 1 state. The forbidden J = 0 to J = 0 (ΔJ = 0) transition is not observed. If it
were observed, the transition would appear as the $\bar{\nu}$ band center transition.
Using algebra, we can obtain the following relations for the wave-numbers (in cm$^{-1}$) of the lines in the R and P branches for the $\nu$ = 0 → 1 transition assuming an initial rotational state $J_o$

$$
\bar{\nu}_R (J_o) = \bar{\nu}_{01}
+ \big( 2 B_e - 3 \alpha_e \big)
+ 2 \big( B_e - 2 \alpha \big) J_o
- \alpha_e J_o^2
$$
and

$$
\bar{\nu}_P (J_o) = \bar{\nu}_{01}
- 2 \big( B_e - \alpha_e \big) J_o
- \alpha_e J_o^2
$$

The rotational quantum number $J_o$ refers to its value in the lower (in this case ν = 0) vibrational state.

**Polynominal curve fitting**

In this lab you will need to do some polynomial curve fitting in numpy.  You will extract the polynomical coefficeints a, b, and c wich will help you solve for various constants in this lab.  Note that when you fit a polynomial function as in the example below, the quadratic coefficient is first, then the linear coefficient, then the constant.

```python
# Generate example data 
x = np.array([0, 1, 2, 3, 4])
y = np.array([3, 8, 12, 8, 2])

# Fit a polynomial curve; 2 means fitting to quadratic
coefficients = np.polyfit(x, y, 2)

# Extract coefficients a, b, and c for the quadratic function y = ax^2 + bx + c
a, b, c = coefficients

# Make up some x values for the fitted curve
x_values = np.linspace(0, 4, 100)

# Calculate the y values using your coefficients
# Note I am using element-by-element math in numpy instead of doing a for loop just to be faster
y_values = a*x_values**2 + b*x_values + c

# Plot the original data and the fitted curve
plt.scatter(x, y, label='Original Data')
plt.plot(x_values, y_values, color='red', label='Fitted Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Curve Fitting')
plt.legend()
plt.grid(True)

plt.show()

print("Coefficients of the fitted quadratic function:\n")
print(f"a = {a:10.8f}")
print(f"b = {b:10.8f}")
print(f"c = {c:10.8f}")

```

**Solving a system of equations**

Solving a system of equations using `sp.solve` is not vastly different than just solving one equation.  You define symbolic variables, use the `sp.Eq` function to define each equation and assemble those into a list.  Then you use `sp.solve` much as you did before.  When you solve a system of equations, the answer is a python dictionary, which may be a data stucture you have not used before.  In a dictionary, each value is labeled with a key.  To access the value, you use the syntax `dictionary_name[key]`.  When you solve a system of equations, the symbolic variables you used are the keys and the answers are the values.  

Consider the following code example which solves the system of equations:
$$
x + y = 1  
$$
$$
2x - y = 4 
$$

The solution is x = 5/3 and  y = -2/3.

```python
#Sovling a system of equations 

# Set up symbolic variables
x, y = sp.symbols("x y")
# Set up the system of equations
# This is just a list of the equations
system_eq = [sp.Eq(x+y, 1), sp.Eq(2*x -y , 4)]

# Solve; the variable ans is a dictionary of the answers
ans = sp.solve(system_eq)
print(ans)

#Now pull out each answer as its own variable and make it a decimal
x_ans = float(ans[x])
y_ans = float(ans[y])
print(f'x: {x_ans}')
print(f'y: {y_ans}')
```

## Calculations

The equations for the frequencies of the P and R branches are second order
polynomials in J. By equating the coefficients from the polynomials to the coefficients in the equations, you can obtain values for most of the desired quantities. For example, for the P branch, we would obtain an equation of the form

$$
\bar{\nu}(J_o) = a J_o^2 + b J_o + c
$$

Comparing this with equation the equation above, we see $c =
\bar{\nu}_{01}$, $b = -2(B_e - \alpha_e)$, and $a = -\alpha_e$.


Once you have obtained $B_e$ you can obtain $r_e$ from the relation

$$
B_e = \frac{h}{8 \pi^2 c \mu r_e^2}
$$


The reduced mass is represented by $μ = m_1 m_2 / (m_1 + m_2) $.  Remember, the $m_1$ and $m_2$ should be the masses of one carbon atom and one oxygen atom, not a whole mole.

The only parameters left to determine are $\bar{\nu}_{0i}$ and $\chi_e$.

Again, using algebra, we can find for any 0 to $i$ vibrational transition
$$
\bar{\nu}_{0i} = i(1 - (i + 1)\chi_e)\bar{\nu}
$$
Writing this for ν = 0 → 1 and ν = 0 → 2 gives

$$
\bar{\nu}_{01} = ( 1 - 2 \chi_e ) \bar{\nu}
$$
and
$$
\bar{\nu}_{02} = 2 ( 1 - 3 \chi_e ) \bar{\nu}
$$

The values of $\bar{\nu}_{0 1}$ and $\bar{\nu}_{0 2}$ are obtained from your fits, and using this information allows you to set up a system of equations to solve for $\bar{\nu}$ and $\chi_e$.

The classical bond dissociation energy, $D_e$ (in cm$^{-1}$) is related
to $\bar{\nu}$ and $\chi_e$ by

$$
\chi_e = \frac{\bar{\nu}}{4 D_e}
$$

You should calculate $D_e$ in cm$^{-1}$, kJ/mol and eV. Finally, the force constant $k_f$ can be determined using

$$
\bar{\nu} = \frac{1}{2 \pi c} \sqrt{ \frac{k_f}{\mu} }
$$

If you are so inclined, you can analyze the R branch in a similar manner and obtain a second set of results.
