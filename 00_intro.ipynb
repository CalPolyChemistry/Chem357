# Introduction to using loops in Python

## General Loop Structures
Recall that we use loops to perform operations many times in python.  You have learned about two types of loops:
- The `for` loop in python performs an operation for every item in a list or array.

- An indexed loop in python, the `range` `for` loop performs an operation a particular number of times.  

## Lists vs. arrays
For loops can be used with python lists and numpy arrays, but numpy arrays have additional useful features.

````python
# Specification of a python list
number_list = [3, 6, 7, 10, 11]

# Print a specific element of the list; counting starts at zero.
print(f"The index 2 element is {number_list[2]}.")

# To do element by element math on a list, you MUST use a for loop
# Something like this doesn't do want you want
print(f'This creates chaos by printing this list twice.')
print(number_list*2)

new_list = []
for number in number_list:
    new_number = number*2
    new_list.append(new_number)
print(f'This is what you actually want, where each element is multiplied by 2.')
print(new_list)

# But if you make the list into a numpy array it DOES work to do element-by-element math
# Convert the list to a numpy array
number_list = np.asarray(number_list)
twice_list = number_list*2
print(f'If you make the list into a numpy array, element-by-elment math does work.')
print(twice_list)
````

## The range for loop

You can also use a range for loop (also called an indexed for loop) to do an operation a particular number of times.

Recall that the syntax is:

```
for index in range(start, end):
  do calculations with index
```
The variable index will be the numbers specified in range.  It is very commom, but not required, to use a single letter variable for the index variable.  If you use a range for loop, you will then index the variable that you want to use.

Just as above, if you want to save the values that you create in a `for` loop, you need to create a blank list or numpy array to save them.  To create a blank numpy array, you use `np.zeros` function.

````python
# Make a list of energy levels; make a numpy array
energy_levels = [1, 2, 3, 4, 5]
energy_levels = np.asarray(energy_levels)

# The zeros function can make an array (1D) or a matrix (2D)
# That is why the five has to be in parenthesis because it could be a matrix
# The index (n) is used to specificy which element of electron_array you are calculating.
electron_array = np.zeros(shape=(5))
for n in range(0, len(energy_levels)):
    electron_array[n] = 2*energy_levels[n]**2   

print(electron_array)
````

## Matricies and higher dimensional arrays

Someting we will do often is work with "higher dimensional arrays" or lists of lists. You could also think of these as a matricies in some contexts.

Just as above, you can make a blank matrix with the `np.zeros` function and then fill in the individual elements by using TWO `range` for loop that counts over the rows and columns.

````python
nrows = 4
ncols = 3
matrix = np.zeros(shape=(nrows, ncols))
print(f'Here is my empty matrix')
print(matrix)

# Let's fill in each matrix element with the sum of the row and column index
for i in range (0, nrows):
    for j in range (0, ncols):
        matrix[i,j] = i+j
print(f'Here is my filled in matrix.')
print(matrix)
````

**Example** Lets try calculating the changes in energy, in eV, for an electron in the hydrogen atom falling from initial n values of 4 through 8 to final n values of 1 through 3. Here is how your code could look using a list of lists.  Recall that the equation is
$$ \Delta E = -kH * \left( \frac{1}{n_f^2} - \frac{1}{n_i^2} \right) $$

````python
# Rhydberg constant in electorn volts
kH = 13.61
ninits = [4,5,6,7,8]
nfinls = [1,2,3]

# Make your empty matrix to store the answers 
# Notice that instead of counting the number of rows and columns I need I make it count for me
dE = np.zeros(shape=(len(ninits), len(nfinls)))

print(f"empty energy transition array")
print(dE)

for i in range (0, len(ninits)):
  for j in range (0, len(nfinls)):
    dE[i][j] = -kH * (nfinls[j]**(-2) - ninits[i]**(-2))

print("\n")
print(f"full energy transition array")
print(dE)

print(f'Print one particular answer')
print(f'For the transition from {ninits[0]} to {nfinls[2]} the energy is {dE[0,2]}.')
````

# Symbolic Calculations in Python with sympy

## 1. Performing Calculus Operations

Often times it is helpful to do symbolic operations with a function. In these operations, the variables remain symbolic, rather than taking on a specific numerical value. For us, taking derivatives, integrating and solving for a variable are the most common operations.

To set up the ideal gas law as a symbolic function and then take the derivative of it, we might use the following code. The first line sets all of our variables as symbolic, the next line defines our expression for p, and then we take the derivative.

The command to take the derivative is diff. It requires two inputs, the function you are taking the derivative of (in this case, p) and what variable you want to take the derivative with respect to (in this case, V). The command pprint just prints your expression in a (slightly) more readable way.

````python
# from sympy import symbols, diff, pprint, solve, integrate

# Remember that we imported sympy in the first cell as "sp"
# so when you want to use as sympy funciton it looks like sp.function_name(stuff)
# The items in the parentheses "stuff" are called the arguemts of the funciton.

p, V, n, R, T = sp.symbols('p V n R T')

# REMEMBER PEMDAS (order of opertations will bite you!)
p = n * R * T / V

derivative = sp.diff(p, V)

# you can display the derivative using the following
display(derivative)

# you can also use sp.pprint() if thats more your style, check it out if youd like
sp.pprint(derivative)
````


# 2. Solving Equations


Another thing we might want to do is set an expression equal to some value (thus creating an equation), and solve for a particular variable.

To do this, use the solve command. The example below sets our function p equal to 1 and solves for V.

If you run this code, you will see that the answer is n*R*T. You can also solve when your function is equal to other variables, rather than a number.

````python
p, V, n, R, T = sp.symbols('p V n R T')

p = n * R * T / V

answer = sp.solve(p - 1, V)

# When you use solve, a list is actyally returned rather than a single value
# If you want the nice pretty expression its the 0th element of the list
print(f"The data type of the varaible answer is: {type(answer)}\n")
print(f"The value of V that solves the equation above is:")
display(answer[0])
````

## 3. Integration

Finally, we might want to integrate our function. The Python command to integrate is `integrate`. There are two types of integrals we might choose to evaluate. An indefinite integral does not have limits. A definite integral has limits.

Let us first integrate p with respect to T (an indefinite integral). If we run this code, the answer is (n*R*T*log(T))/V, as expected (no constant of integration is included). Notice that for an indefinite integral, integrate only needs two arguments, the expression to integrate and the variable we want to integrate with respect to.

````python
# Note that we are defining the symbols newly in each cell,
# you don't have to do that, we're just doing it here for clarity

p, V, n, R, T = sp.symbols('p V n R T')
p = n * R * T / V
integral = sp.integrate(p, T)

print(f"The indefinite integral evaluates to:")
display(integral)
````

Now, let us do a definite integral calculation. In this case integrate requires four arguments, the expression, the variable to integrate with respect to, the lower limit, and the upper limit. To integrate p with respect to T and evaluate it from T1 = 0 to T2 = 1, you would use the following code to generate the answer.

````python


p, V, n, R, T = sp.symbols('p V n R T')
p = n*R*T/V
integral = sp.integrate(p, (T, 0, 1))

print(f"The definate integral with bounts from T=0 to T=1 evaluates to:")
display(integral)
````

# 4. From symbolic to numeric

This will be reviewd throught the quarter, but the vast majority of computational science isn't concerned with using comupers to find analytical symbolic expression, but rather using computers to do numerical calcualtions. 

`sympy` can turn a symbolic expression into a numerical expression via sympy.lambdify() to actually give you values, this will be useful for plotting and a wide variety of other applicaoitns.

````python
# Define symbolic variables
p, V, n, R, T = sp.symbols('p V n R T')

# Define the ideal gas law expression
ideal_gas_law = n*R*T/V

# Convert the ideal gas law expression to a numerical function that
# will take numerical values of p, V, n R, and T.
numerical_ideal_gas_law = sp.lambdify((V, n, R, T), ideal_gas_law)

# Example numerical values
n_val = 2.0
R_val = 8.314  # J/(mol*K)
T_val = 300.0  # K
V_val = 0.0224  # m^3

# Calculate the pressure using the numerical expression
pressure = numerical_ideal_gas_law(V=V_val, n=n_val, R=R_val, T=T_val)
print(f"Pressure in Pa: {pressure:.4f}")

# note that {pressure:.4f} tells print that you'd like a max of
# 4 decimals places to be printed try {pressure:.6f} and
# {pressure:e} to see how things change
````
