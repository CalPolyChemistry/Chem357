## Introduction

A blackbody is a theoretical object that absorbs all wavelengths of radiation. The
radiant energy density, $\rho$, can be defined in terms of the frequency $\nu$.

$E = \int \rho(\nu) d \nu = \int \frac{8 \pi h \nu^3}{c^3} \frac{1}{e^{h\nu/kT} - 1} d\nu$

Note that, since $c = \lambda \nu$, the radiant energy could also be found by integrating over wavelength.

$E = \int \rho(\lambda) d\lambda$

## General Programming Tips

a. For symbolic manipulation you may want to assume that certian variables (such as temperature) are always positive. This can be acomplished using sympy (sp), for example:

T = sp.symbols('T', positive=True)

b. In many programming languages, log means “natural log”. Base 10 log is log10.

c. Sympy can be used to calcualte valeus in additon to just symbolic manipulation. This can be acomplished using the `sp.lambdify` function.  If you have a sympy symbolic symbol (like `sp.pi`) then you have to tell the lambidy function to use numpy to convert these to numerical symbols.  An example is shown in the code block below.

d. If you want to set a sympy expression equal to a value and then solve it, the syntax is: 

`answer = sp.solve(sp.Eq(sympy_expression, value), variable_to_solve_for)`


```python
# Relevant code examples

#Symbolic function to find the area of a circle
r = sp.symbols('r')
#The radius is a distance so it is always positive
r = sp.symbols('r', positive=True)
area_sym = sp.pi * r**2
display(area_sym)

#Now make this into a function I can actually call to get a number
#The addition of 'numpy' at the end changes any symbolic sympy symbols to their numpy numerical version
area_num = sp.lambdify([r], area_sym, 'numpy')

#Now I could call the function and get an numerical answer
radius = 3
area = area_num(r=radius)
print(F'The area of a circle with a radius of {radius} is {area:.3f}.')

#Now I could call this function within a for loop to calculate the area for a list of radii
radius_list = [1, 3, 5, 7]
#Recast to a numpy array
radius_array = np.asarray(radius_list)
#Make a blank array for the areas
area_array = np.zeros(shape=(len(radius_array)))

# Use the area_num function to calculate the areas
for i in range(0,len(radius_array)):
    area_array[i] = area_num(radius_array[i])

print(area_array)

# Now you could set the SYMPY function equal to a number and then solve for the radius
# This would find the radius of a circle with area 10
# Remember the output of solve is a list, so the answer is element zero
answer = sp.solve(sp.Eq(area_sym, 10) , r)
print(answer[0])
#If it doesn't make the answer a decimal, you might have to recast to a float
answer = float(answer[0])
print(answer)
```
